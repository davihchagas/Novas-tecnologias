# Exercício 1 - Cadastro e análise simples de veículos

Uma concessionária deseja registrar alguns veículos disponíveis para venda.
Crie um programa em Python que crie uma lista de veículos.

Cada veículo deve ser representado por um dicionário contendo:
- marca
- modelo
- preço
- ano

Exemplo de estrutura esperada:

```
veiculo = {
    "marca": "Toyota",
    "modelo": "Corolla",
    "preco": 120000,
    "ano": 2022
}
```

O programa deve:

- Adicionar pelo *menos 5 veículos* à lista.
- Exibir todos os veículos cadastrados.
- Calcular o *preço médio dos veículos*.
- Mostrar quais *veículos custam acima da média*.

### Saída esperada (exemplo)
```
Preço médio dos veículos: 85000
Veículos acima da média:
Honda Civic - 110000
Toyota Corolla - 120000
```


````python
# Solução do exercício 1

veiculo1 = {"marca": "Fiat", "modelo": "Uno", "preco": 50000, "ano": 2020}
veiculo2 = {"marca": "Ford", "modelo": "F-150", "preco": 143900, "ano": 2023}
veiculo3 = {"marca": "RAM", "modelo": "350", "preco": 343900, "ano": 2023}
veiculo4 = {"marca": "Hyundai", "modelo": "Creta", "preco": 133900, "ano": 2022}
veiculo5 = {"marca": "Toyota", "modelo": "Corolla Cross", "preco": 219890, "ano": 2026}

def criar_lista(*args):
  veiculos = []
  for arg in args:
    veiculos.append(arg)
  return veiculos

veiculos = criar_lista(veiculo1, veiculo2, veiculo3, veiculo4, veiculo5)
qtd_veiculos = 0
total = 0

print("Veículos cadastrados\n")
for veiculo in veiculos:
  qtd_veiculos += 1
  total += veiculo["preco"]

  print(f"Marca: {veiculo["marca"]} \n"
        f"Modelo: {veiculo["modelo"]} \n"
        f"Preço: {veiculo["preco"]} \n"
        f"Ano: {veiculo["ano"]} \n")

media = total/qtd_veiculos

print(f"Preço médio: {media}\n")
print("Veículos acima da média\n")
for veiculo in veiculos:
  if veiculo["preco"] > media:
    print(f"Marca: {veiculo["marca"]} \n"
        f"Modelo: {veiculo["modelo"]} \n"
        f"Preço: {veiculo["preco"]} \n"
        f"Ano: {veiculo["ano"]} \n")



````

# Exercício 2 - Sistema simples de financiamento

Uma concessionária oferece financiamento de veículos.

Crie uma função chamada:

```
calcular_financiamento(preco, entrada, parcelas)
```

A função deve calcular o valor financiado.

```
valor_financiado = preco - entrada
```

Considerar uma taxa de juros simples de 1.5% ao mês.

Fórmula simplificada:
```
valor_total = valor_financiado * (1 + taxa * parcelas)
parcela = valor_total / parcelas
```
A função deve retornar o valor da parcela.


### Crie um programa que:

- Defina o preço de um veículo
- Pergunte ao usuário: valor de entrada e número de parcelas
- Use a função para calcular o financiamento
- Exiba o valor da parcela
- Impedir parcelas maiores que 72 meses
- Mostrar uma mensagem caso isso ocorra (Número máximo de parcelas é 72.)

### Exemplo de saída

```
Preço do veículo: 80000
Entrada: 20000
Parcelas: 48
Valor da parcela: 1785.50
```


````python
# Solução do exercício 2

def calcular_financiamento(preco, entrada, parcelas):
  if parcelas > 72:
    return "Número máximo de parcelas é 72"
  valor_financiado = preco - entrada
  valor_total = valor_financiado * (1 + 0.015 * parcelas)
  parcela = valor_total / parcelas
  return parcela

preco_veiculo = 80000

valor_entrada = float(input("Digite o valor de entrada:"))
parcelas = int(input("Digite a quantidade de parcelas:"))

financiamento = calcular_financiamento(preco_veiculo, valor_entrada, parcelas)

texto = (f"Preço do veículo: {preco_veiculo}\n"
         f"Entrada: {valor_entrada}\n"
         f"Parcelas: {parcelas}\n"
         f"Valor da parcela: {financiamento}\n"
        )

print(texto)

````
